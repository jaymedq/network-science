```python
import networkx as nx

#Load GraphMissingEdges
G = nx.read_gml("../GraphMissingEdges.gml")
```

```python
def compute_centrality(centrality_callback):
    # Compute degree centrality
    degree_centrality = centrality_callback(G)

    # Sort nodes by degree centrality and get top 50
    top_50_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:50]

    # Print top 50 places with their categories
    print("Top 50 places according to centrality:")
    for node, centrality in top_50_degree_centrality:
        print(f"Place: {G.nodes[node].get('name','Empty')}")
        print(f"Centrality: {centrality}")
        print(f"Categories: {G.nodes[node]['categories']}\n")
```

## 3- Explique resumidamente a ideia de Hubs e Authorities no contexto de métricas de centralidade de rede. 

No contexto de métricas de centralidade de rede, **Hubs** e **Authorities** são conceitos que se referem à importância de um nó na rede. Eles são interdependentes e foram inicialmente utilizados para modelar páginas da web.

- **Hubs**: Um bom hub é um vértice que aponta para muitos vértices com alto coeficiente de Authority. Ou seja, um hub é um nó que tem muitas conexões de saída para nós de alta qualidade.

- **Authorities**: Um bom Authority é um vértice que é apontado por muitos bons Hubs. Ou seja, uma autoridade é um nó que tem muitas conexões de entrada de nós de alta qualidade.

O algoritmo HITS (Hyperlink-Induced Topic Search) é frequentemente usado para calcular os coeficientes de Hub e Authority. Esses conceitos são particularmente úteis em redes direcionadas e acíclicas, onde a importância (ou centralidade) de um nó pode ser avaliada com base em quão bem ele atua como um hub (apontando para outros nós) ou uma autoridade (sendo apontado por outros nós). Essas métricas ajudam a identificar os nós mais importantes em uma rede.
