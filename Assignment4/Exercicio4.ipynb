```python
import networkx as nx
import pandas as pd

#Load GraphMissingEdges
G = nx.read_gml("../GraphMissingEdges.gml")
```

a- Itendifique comunidades usando o método Leiden ou Louvain.

```python
import community as community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt
#first compute the best partition
partition = community_louvain.best_partition(G)
print(partition)
print(len(partition.values()))
# draw the graph
pos = nx.spring_layout(G)
# color the nodes according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

b- Identifique comunidades utilizando o k-clique. Utilize k=4.

```python
k_clique_communities = list(nx.algorithms.community.k_clique_communities(G, 4))

for member in k_clique_communities:
    print(sorted(list(member)))
```

c- Traduza os ids dos nós em cada cluster pelos nomes de suas categorias (referentes aos locais que 
eles se referem). Os resultados parecem fazer sentido? Mostre um exemplo que ilustra a sua 
resposta.

```python
category_mapping = pd.read_csv("../KaggleChallenge/predio-de-links-utfpr-2024-1/categories.csv")
category_dict = category_mapping.set_index("CategoryId")["names"].to_dict()

parsed_partitions = {}
for key, value in partition.items():
    categories = G.nodes[key]["categories"].split(',')
    try:
        if any(categories):
            categories_labels = [category_dict[int(category)] for category in categories] if categories else categories
        else:
            categories_labels = key
    except Exception as e:
        print(G.nodes[key])
        print(categories)
        raise e
    parsed_name = '_'.join(categories_labels) if any(categories) else key
    parsed_partitions.update({parsed_name:value})

grouped_dict = {}
for key, value in parsed_partitions.items():
    if value not in grouped_dict.keys():
        grouped_dict[value] = [key]
    else:
        grouped_dict[value].append(key)
```

```python
# Exibir os resultados
print("Comunidades identificadas usando Louvain:")
for key, value in sorted(grouped_dict.items()):
    print(f"Categorias da Comunidade {key}: {value}")
```

```python
# Carregar o mapeamento de categoria
category_mapping = pd.read_csv("../KaggleChallenge/predio-de-links-utfpr-2024-1/categories.csv")
category_dict = category_mapping.set_index("CategoryId")["names"].to_dict()

# Função para converter IDs de categoria em nomes de categoria
def parse_categories(categories):
    categories_labels = [category_dict[int(category)] for category in categories.split(',')] if categories else []
    return categories_labels

# Parser para as comunidades identificadas pelo método k-clique
k_clique_communities_names_parsed = []
for community in k_clique_communities:
    community_names = []
    for node in community:
        node_categories = G.nodes[node]["categories"]
        parsed_categories = parse_categories(node_categories)
        community_names.extend(parsed_categories)
    k_clique_communities_names_parsed.append(community_names)
```

```python

# Exibir as comunidades identificadas pelo método k-clique com os nomes das categorias
print("\nComunidades identificadas usando k-clique (k=4):")
for i, community in enumerate(k_clique_communities_names_parsed):
    print(f"Categorias da Comunidade {i+1}: {community}")
```
